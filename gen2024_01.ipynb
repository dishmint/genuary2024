## Model

```
particle = <|
	"ID" -> CreateUUID[],
	"Position" -> {0,0},
	"Velocity" -> {0,0},
	"Acceleration" -> {0,0},
	"Color" -> Red
|>;
```

```
Graphics[{particle["Color"], Disk[particle["Position"], 5]}]
```

```
newparticle[] := <|
	"ID" -> CreateUUID[],
	"Position" -> RandomReal[{-9,9},2],
	"Velocity" -> {0,0},
	"Acceleration" -> {0,0},
	"Color" -> RandomColor[]
|>;
```

```
tenTParticles=Table[newparticle[],10000];
```

```
Graphics[(MapAt[Disk[#,1]&, 2]@*Lookup[{"Color","Position"}])/@tenTParticles, PlotRange -> {-10,10}]
```
